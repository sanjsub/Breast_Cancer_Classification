```python
import pandas as pd
```

```python
train = pd.read_csv('DBT_train.csv')
train.head()
```

```python
for i in range(len(train)):
    path = train['descriptive_path'].iloc[i]
    new_path = path[:-13] + 'NA-' + path[-13:-3] + 'jpg'
    train['descriptive_path'].iloc[i] = new_path
train
```
