```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
original_df = pd.read_csv('Train_Labels.csv')
print(len(original_df))
original_df.head()
```

```python
df = original_df[original_df['Actionable']!=1]
print(len(df))
print('Number of actionable labels removed:', len(original_df)-len(df))
df.head()
```

```python
df.loc[df['Benign']==1,'Label'] = 1
df.loc[df['Cancer']==1,'Label'] = 1
df.loc[df['Normal']==1,'Label'] = 0
df['Label'] = df['Label'].astype(int)
df
```

```python
# Make sure that mapping worked
print(df['Benign'].sum() + df['Cancer'].sum())
print(df['Label'].sum())
df
print(df['Benign'].sum())
print(df['Cancer'].sum())
print(df['Normal'].sum())
print(200/len(df))
print(df['Label'].sum() / len(df))
```

```python
unique_patients = df.drop(['View','Actionable'], axis=1)
unique_patients = unique_patients.drop_duplicates()
print(len(unique_patients))
unique_patients
```

```python
print('Number of Normal Patients:', len(unique_patients[unique_patients['Normal']==1]))
print('Number of Benign Patients:', len(unique_patients[unique_patients['Benign']==1]))
print('Number of Cancer Patients:', len(unique_patients[unique_patients['Cancer']==1]))
```

```python
unique_patients = unique_patients.drop(['Normal','Benign','Cancer'], axis=1)
unique_patients.head()
```

```python
X = unique_patients.drop('Label', axis=1)
y = unique_patients['Label']

X_nontest, X_test, y_nontest, y_test = train_test_split(X, y, test_size=0.1, random_state=1, shuffle=True, stratify=y)
print(y_test.sum()/(len(X_test)+len(y_test)))
X_train, X_val, y_train, y_val = train_test_split(X_nontest, y_nontest, train_size=7/9, random_state=1, shuffle=True, stratify=y_nontest)
print(y_train.sum()/(len(X_train)+len(y_train)))
print(y_val.sum()/(len(X_val)+len(y_val)))

print('train:', len(X_train)/len(df), len(X_train))
print('val:', len(X_val)/len(df), len(X_val))
print('test:', len(X_test)/len(df), len(X_test))

print(len(X_train)+len(X_val)+len(X_test))
```

```python
# Combine back into one dataframe
X_train['Label'] = y_train
X_val['Label'] = y_val
X_test['Label'] = y_test
```

```python
# Make sure there is no patient overlap between the train/val/test splits
patients_train = set(X_train['PatientID'])
patients_val = set(X_val['PatientID'])
patients_test = set(X_test['PatientID'])

print(patients_train.intersection(patients_val, patients_test))
print(patients_val.intersection(patients_train, patients_test))
print(patients_test.intersection(patients_train, patients_val))
```

```python
X_train_view = pd.merge(df, X_train, how='inner', on=['PatientID','StudyUID'])
X_train_view = X_train_view.drop(['Normal','Actionable','Benign','Cancer','Label_x'], axis=1)
X_train_view = X_train_view.rename({'Label_y':'Label'}, axis=1)
X_train_view
```

```python
X_val_view = pd.merge(df, X_val, how='inner', on=['PatientID','StudyUID'])
X_val_view = X_val_view.drop(['Normal','Actionable','Benign','Cancer','Label_x'], axis=1)
X_val_view = X_val_view.rename({'Label_y':'Label'}, axis=1)
X_val_view
```

```python
X_test_view = pd.merge(df, X_test, how='inner', on=['PatientID','StudyUID'])
X_test_view = X_test_view.drop(['Normal','Actionable','Benign','Cancer','Label_x'], axis=1)
X_test_view = X_test_view.rename({'Label_y':'Label'}, axis=1)
X_test_view
```

```python
print(len(X_train_view)+len(X_val_view)+len(X_test_view))
```

```python
paths = pd.read_csv('Train_File_Paths.csv')
print(len(paths))
paths.head()
```

```python
train = pd.merge(X_train_view, paths, how='inner', on=['PatientID','StudyUID', 'View'])
print(len(train))
print(len(train[train['Label']==0]))
print(train['Label'].sum())
print('Ratio:', train['Label'].sum() / len(train[train['Label']==0]))
train.head()
```

```python
val = pd.merge(X_val_view, paths, how='inner', on=['PatientID','StudyUID','View'])
print(len(val))
print(len(val[val['Label']==0]))
print(val['Label'].sum())
print('Ratio:', val['Label'].sum() / len(val[val['Label']==0]))
val.head()
```

```python
test = pd.merge(X_test_view, paths, how='inner', on=['PatientID','StudyUID','View'])
print(len(test))
print(len(test[test['Label']==0]))
print(test['Label'].sum())
print('Ratio:', test['Label'].sum() / len(test[test['Label']==0]))
test.head()
```

```python
# The difference is the number of actionable labels that were removed in the beginning
print(len(paths))
print(len(train) + len(val) + len(test))
```

```python
print(train['Label'].sum()/len(train))
print(val['Label'].sum()/len(val))
print(test['Label'].sum()/len(test))
```

```python
train.to_csv('DBT_train.csv', index=False, header=True)
val.to_csv('DBT_val.csv', index=False, header=True)
test.to_csv('DBT_test.csv', index=False, header=True)
```
